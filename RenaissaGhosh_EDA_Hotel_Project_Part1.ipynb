```python
# Task 1: Convert messy, unclean data into clean data by performing missing value treatment
mean_a = df["agent"].mean()
df["agent"] = df["agent"].fillna(mean_a)

mean_c = df["company"].median()
df["company"] = df["company"].fillna(mean_c)

df
```

```python
conti = df.select_dtypes(include="number")
for i in conti.columns:
    print(df[i].value_counts())
```

```python
# Task 2: Booking Pattern Analysis

# Convert arrival date if not already present
df['arrival_date'] = pd.to_datetime(df['arrival_date_year'].astype(str) + '-' + df['arrival_date_month'] + '-' + df['arrival_date_day_of_month'].astype(str))

# I. Peak booking periods
monthly_bookings = df.groupby(df['arrival_date'].dt.to_period('M')).size()
monthly_bookings.plot(kind='bar', figsize=(6, 4))
plt.title('Monthly Booking Counts')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.tight_layout()
plt.show()

# II. Lead time vs cancellations
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x='is_canceled', y='lead_time')
plt.title('Lead Time vs Cancellation')
plt.xlabel('Cancelled (0=No, 1=Yes)')
plt.ylabel('Lead Time')
plt.tight_layout()
plt.show()

# III. Booking patterns by month, week, day of the week
df['day_of_week'] = df['arrival_date'].dt.day_name()
weekly_counts = df['day_of_week'].value_counts().reindex(['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
plt.figure(figsize=(6, 4))
weekly_counts.plot(kind='bar')
plt.title('Bookings by Day of the Week')
plt.xlabel('Day')
plt.ylabel('Number of Bookings')
plt.tight_layout()
plt.show()
```

```python
# Task 3: Booking Cancellation Analysis

# I. Factors influencing cancellations
corr = df[['lead_time', 'adr', 'is_canceled']].corr()
sns.heatmap(corr, annot=True)
plt.title('Correlation with Cancellations')
plt.tight_layout()
plt.show()

# II. Trends in cancellations over time
cancellations_over_time = df[df['is_canceled']==1].groupby(df['arrival_date'].dt.to_period('M')).size()
cancellations_over_time.plot(figsize=(6, 4))
plt.title('Cancellations Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Cancellations')
plt.tight_layout()
plt.show()

# III. Identify high-risk bookings
high_risk = df[df['lead_time'] > df['lead_time'].median()]
print("High-risk bookings likely to cancel (lead time above median):", high_risk['is_canceled'].mean())
```

```python
# Task 4: Customer Behavioral Segmentation

# I. Patterns by customer segments
df['customer_type'] = df['customer_type'].astype('category')
segment_summary = df.groupby('customer_type').agg({'lead_time':'mean', 'total_of_special_requests':'mean'})
print(segment_summary)

# II. Effective marketing channels
channel_cancellation_rate = df.groupby('distribution_channel')['is_canceled'].mean().sort_values()
plt.figure(figsize=(10, 6))
channel_cancellation_rate.plot(kind='bar')
plt.title('Cancellation Rate by Distribution Channel')
plt.xlabel('Channel')
plt.ylabel('Cancellation Rate')
plt.tight_layout()
plt.show()

# III. Customize marketing strategies
high_special_request_segments = df.groupby('customer_type').agg({'total_of_special_requests':'mean', 'is_canceled':'mean'})
print("Segments to target with special offers:\n", high_special_request_segments)
```

```python
# Task 5: Revenue Management

# I. ADR over time
adr_trend = df.groupby(df['arrival_date'].dt.to_period('M'))['adr'].mean()
adr_trend.plot(figsize=(6, 4))
plt.title('Average Daily Rate (ADR) Over Time')
plt.xlabel('Month')
plt.ylabel('ADR')
plt.tight_layout()
plt.show()

# II. Pricing strategies
high_adr = df[df['adr'] > df['adr'].quantile(0.75)]
print("Top 25% bookings by ADR, mean ADR:", high_adr['adr'].mean())

# III. Room types with higher ADR
room_adr = df.groupby('reserved_room_type')['adr'].mean().sort_values(ascending=False)
room_adr.plot(kind='bar')
plt.title('ADR by Room Type')
plt.xlabel('Room Type')
plt.ylabel('ADR')
plt.tight_layout()
plt.show()
```

```python
# Task 6: Operational Efficiency

# I. Effectiveness of distribution channels
channel_eff = df.groupby('distribution_channel').size().sort_values(ascending=False)
plt.figure(figsize=(6, 4))
channel_eff.plot(kind='bar')
plt.title('Bookings by Distribution Channel')
plt.xlabel('Channel')
plt.ylabel('Bookings')
plt.tight_layout()
plt.show()

# II. Booking changes vs operations
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='booking_changes', hue='is_canceled')
plt.title('Booking Changes and Cancellation Status')
plt.xlabel('Number of Changes')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# III. Waiting list duration vs cancellation
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x='is_canceled', y='days_in_waiting_list')
plt.title('Waiting List Duration vs Cancellation')
plt.xlabel('Cancelled (0=No, 1=Yes)')
plt.ylabel('Days in Waiting List')
plt.tight_layout()
plt.show()
```

```python
# Task 7: Loyalty Programs and Repeat Business

# I. Behavior of repeated guests
repeat_stats = df.groupby('is_repeated_guest').agg({'adr':'mean', 'lead_time':'mean'})
print("Behavior of repeated vs. non-repeated guests:\n", repeat_stats)

# II. Loyalty programs and special requests
loyalty_special = df.groupby('is_repeated_guest')['total_of_special_requests'].mean()
loyalty_special.plot(kind='bar')
plt.title('Average Special Requests by Guest Type')
plt.xlabel('Repeated Guest (0=No, 1=Yes)')
plt.ylabel('Average Special Requests')
plt.tight_layout()
plt.show()

# III. Factors for repeat business
sns.boxplot(data=df, x='is_repeated_guest', y='lead_time')
plt.title('Lead Time for Repeated vs. Non-Repeated Guests')
plt.xlabel('Repeated Guest (0=No, 1=Yes)')
plt.ylabel('Lead Time')
plt.tight_layout()
plt.show()
```
